```python
def dresser_clean(text):
    text = text.replace('</p>, <p>', '<\n\n> ')
    text = text.replace('[<p>', '')
    text = text.replace('</p>]', '')
    text = text.replace('<\n\n> —<\n\n>', '<\n\n>')
    text = text.
    
    return text
```
